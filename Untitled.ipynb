```python
PATH = "data/"


data = dict()

for filename in os.listdir(PATH):
    country = dict()
    with open(PATH+filename) as f:
        line = f.readline()
        while line:
            if line[0] == '"':
                row = line.split("\t")[:-1]
                country[row[0][1:-1].lower()] = row[1:]
            line = f.readline()

    data[filename.split('.')[0]] = country
    
data
```

```python
columns = [
    "Value exported in 2021 (USD thousand)",
    "Trade balance 2021 (USD thousand)",
    "Share in Australia's exports (%)",
    "Growth in exported value between 2017-2021 (%, p.a.)",
    "Growth in exported value between 2020-2021 (%, p.a.)",
    "Ranking of partner countries in world imports",
    "Share of partner countries in world imports (%)",
    "Total imports growth in value of partner countries between 2017-2021 (%, p.a.)",
    "Average distance between partner countries and all their supplying markets (km)",
    "Concentration of all supplying countries of partner countries",
    "Average tariff (estimated) faced by Australia (%)"
]
```

```python
def generate_dataframe_from(data, column_i):
    rows = []
    for exporter_k in data:
        exporter = data[exporter_k]
        row = np.zeros(len(data))
        i = 0
        for to in data:
            if to in exporter and exporter[to][column_i]:
                row[i] = float(exporter[to][column_i])
            i += 1
        rows.append(row)
    return pd.DataFrame(np.array(rows), columns = data.keys(), index = data.keys())
```
